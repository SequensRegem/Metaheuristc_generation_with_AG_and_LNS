```python
import numpy as np
import random
```

```python
individual = np.random.randint(0,2,7)
```

```python
def removal_function(individual,n_remove):
    individual = list(individual)
    length = len(individual)
    indexes = random.sample(range(0,length),n_remove)
    values_list = []
    for _ in indexes:
        values_list.append(individual[_])

    return(values_list,indexes)
```

```python
def insertion_function(values_list,indexes,individual):
    f = lambda x: 1-x
    values_transformed = [f(x) for x in values_list]
    for index, _ in enumerate(indexes):
        individual[_] = values_transformed[index]
        
    return(individual,values_transformed)
    
```

```python
def fitness_function(individual):
    weights = [5, 25, 75, 50, 20, 30, 40]
    values = [50, 100, 500, 200, 50, 80, 20]
    value_total = 0
    weight_total = 0
    
    for index, active in enumerate(individual):
        if active:
            value_total += values[index]
            weight_total += weights[index]
            
    if weight_total > 100:
        value_total = 0
        return(value_total)
    else:
        return(value_total)
```

```python
def implementation_function(individual,n_remove,k):
    best_fitness = -1
    best_individual = []
    
    for _ in range(k):
        values_list,indexes = removal_function(individual,n_remove)
        individual,valeus_list = insertion_function(values_list,indexes,individual)
        fitness = fitness_function(individual)
        
        if fitness > best_fitness:
            best_individual = individual
            best_fitness = fitness
        else:
            individual,values_list = insertion_function(values_list,indexes,individual)
            values_list = []

    return(best_fitness,best_individual)
```

```python
best_fitness,best_individual = implementation_function(individual,3,100)
```
